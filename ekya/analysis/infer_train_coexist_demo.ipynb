```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size':20})
```

```python
infer_accuracy_timestamps = [0, 13.5, 26.9, 58.56, 89.07, 120.8, 146.15, 159.57]
infer_accuracy_data = [0.10, 0.021,0.020, 0.018, 0.022, 0.021, 0.069, 0.071]
infer_latency_data = [13.65/5, 13.15/5, 31.75/5, 30.51/5, 31.1/5, 14.92/5, 13.12/5, 13.28/5]

#infer_latency_timestamps = [13.5, 26.9, 58.56, 89.07, 120.8]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(infer_accuracy_timestamps, infer_accuracy_data, marker='*')
ax1.set_ylabel('infer_accuracy (%)')

ax2 = ax1.twinx()
ax2.plot(infer_accuracy_timestamps, infer_latency_data, 'r-', marker='o')
ax2.set_ylabel('per sample infer_latency (ms)', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')
    
plt.xlabel("Wall time")
```
