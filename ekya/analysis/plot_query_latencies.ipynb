```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
import glob
files = (glob.glob("./tmp/tmp2/*.csv"))
```

```python
plt_data = []
min_t = 999999999999999999999
for file in files:
    print(file)
    with open(file, 'r') as f:
        reader = csv.reader(f)
        data = list(reader)
        x=[float(d[0]) for d in data]
        y=[float(d[1]) for d in data]
        min_t = min(min_t, min(x))
    plt_data.append([x,y])
    #plt.plot(x,y)
```

```python
for x,y in plt_data:
    time_data = np.array(x)-min_t
    plt.plot(time_data,y)
    plt.axvline(x=max(time_data))
plt.ylim([0,0.5])
plt.xlabel("Time")
plt.ylabel("Query latency")
```
